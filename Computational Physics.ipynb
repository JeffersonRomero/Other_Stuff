```python
# We are not going find the root of a fuction (when fuction=0) by using the Bisection Method

# We import our libraries
import numpy as np
import matplotlib.pyplot as plt

# We define an easy fuction which is eaqual to zero just once (you will see it in the graphic)
def f(x):
    return x ** 3 - 2 * x + 2

# We implement our method
def root_bisection(f, a, b, tolerance=1.0e-6):
    dx = abs(b - a)
    while dx > tolerance:
        x = (a + b) / 2.0
        if f(a) * f(x) < 0:  
            b = x
        else:  
            a = x  
        dx = abs(b - a)
    return x

# We create our values range
x_values = np.linspace(-2, 2, 100)
# We calculate the values of our fuction
f_values = [f(x) for x in x_values]

# We graph our fuction like this:
plt.plot(x_values, f_values)
plt.axhline(0, color='gray', linewidth=0.5)  
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()

""""We ask the user to see the graph and enter the nearest value at the left 
    and the nearest value at the right where f(x)=0"""

a = float(input("Enter the nearest value at the left that you notice where F(x)=0  (a): "))
b = float(input("Enter the nearest value at the right that you notice where F(x)=0  (b): "))

# We found the root of the equation f(x)=0
root = root_bisection(f, a, b)

# We print the results
print("The root of the fuction is f(x)=0: ", root)
```

```python
# We will be solving now Numeric Integration

# We import our libraries
import numpy as np
import matplotlib.pyplot as plt

def int_simple(f, dx): # This is the simple method
    
    return dx * sum(f[0:-1])  # The sum() function is built-in, and does exactly what you would expect.

# We define de sin fuction
def f(x):
    return np.sin(x)

# These are the integral's parameters
N = 100  # number of divisions
a = 0.0  # lower limit of integration
b = np.pi  # upper limit of integration

# Integration intervals
interval = (b - a) / float(N)

# We calculate the values for x
x = np.linspace(a, b, N+1)

# We calculate the values for f(x)
values_function = [f(value) for value in x]

# We calculate the aproximation of the integral
aproximation_integral = int_simple(values_function, interval)

# We show the fuction and the rectangles of the integral aproximation to see what we are doing
plt.plot(x, values_function, label='Sin f(x)')
plt.bar(x[:-1], values_function[:-1], width=interval, align='edge', color='lightblue', edgecolor='black', label='Rectangles')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Integral approximation using rectangles')
plt.legend()
plt.grid(True)
plt.show()

# We print the result
print("The approximate value of the integral is :", aproximation_integral)
```

```python
# We will now do the same but simpler using a library who uses a Trapezoid Method

# We import our libraries
import numpy as np
import matplotlib.pyplot as plt

# We define de sin function
def f(x):
    return np.sin(x)

# These are the integral's parameters
N = 100  # number of divisions
a = 0.0  # lower limit of integration
b = np.pi  # upper limit of integration

# We calculate the values for x
x = np.linspace(a, b, N+1)

# We calculate the values for f(x)
values_function = f(x)

# We calculate the approximation of the integral using np.trapz
approximation_integral = np.trapz(values_function, x)

# We show the function and the trapezoids of the integral approximation
plt.plot(x, values_function, label='Sin f(x)')
plt.fill_between(x, values_function, color='lightblue', alpha=0.5, label='Trapezoids')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Integral approximation using trapezoids')
plt.legend()
plt.grid(True)
plt.show()

# We print the result
print("The approximate value of the integral is:", approximation_integral)
```

```python
# We will solve the differential equation of a Simple Harmonic Oscillator

# We import our libraries

import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['text.usetex'] = True  # This activates LaTeX when using Matplotlib
import matplotlib.pyplot as plt
import numpy as np

# We define our parameters given the problem
N = 1000  # Number of steps to take
xo = 0.0  # initial position (spring unstretched)
vo = 0.0  # initial velocity
tau = 3.0  # total time for the simulation (s)
dt = tau / float(N - 1)  # time step
k = 3.5  # Spring constant (N/m)
m = 0.2  # Mass (kg)
g = 9.8  # Gravity (m/s^2)

"""
We create an Nx2 array for storing the results of our calculations. 
Each row with 2 elements will be used for the state of the system at one instant, 
and each instant is separated by time dt. 
The first element in each row will be position, the second velocity.
                                                                    """
y = np.zeros((N, 2))
y[0, 0] = xo  # We set the initial state
y[0, 1] = vo

"""
We define the differential ecuation of the simple harmonic oscillator (SHO) like 
dx^2/dt = -k/m x - g, or dv/dt=-k/m x-g because dx/dt=v, so we set the next function
"""

def SHO(state, time):
    
    g0 = state[1]
    g1 = -k / m * state[0] - g
    return np.array([g0, g1])

# We make the calculation as follow using the Euler's Method:
for j in range(N - 1):
    y[j + 1] = y[j] + dt * SHO(y[j], j * dt)

# We use the data stored in y to generate the data for x and v (lists) like this:
time = np.linspace(0, tau, N)
xdata = [y[j, 0] for j in range(N)]
vdata = [y[j, 1] for j in range(N)]

# We show the results
plt.plot(time, xdata, label=r'Position ($x$)')
plt.plot(time, vdata, label=r'Velocity ($v$)')
plt.xlabel(r'$t (s)$', fontsize=14)
plt.legend()
plt.title(r'$Simple Harmonic Oscillator$', fontsize=16)
plt.title(r'$\ddot{x} = -\frac{k}{m} x - g$', fontsize=16) # This shows the equation we solved!
plt.show()
```

```python
# We are going to show a Poincar Section of the Duffing Oscillator 

# We import our Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint # To solve differential equations

"""
The equation for this Oscillator is: 
x + x + x + x =  cos(t)
Where: =alpha, =beta, =delta, =gamma, =omega 
and these are the parameter of the Duffing Oscillator given in the problem
                                                        """
alpha = 1
beta = -1
delta = 0.2
gamma = 0.3
omega = 1

# This is how we define the fuction for the Differential Equations
def duffing(x, t):
    dxdt = np.zeros_like(x)
    dxdt[0] = x[1]
    dxdt[1] = -delta * x[1] - beta * x[0] - alpha * x[0]**3 + gamma * np.cos(omega * t)
    return dxdt

# We generate the time interval that goes from o to 1000, with a total of 100000 points
t = np.linspace(0, 1000, 100000)

# We set the initial conditions of the oscillator
x0 = [0.1, 0]

"""is is how we solve the differential equations using odeint 
#hich numerically integrates the system of differential equations defined in duffing"""
x = odeint(duffing, x0, t)

""" the next loop is represented the Poincar Section which generates 
and stores the values for the postion and velocity"""
poincare_x = []
poincare_v = []

for i in range(len(x)-1):
    if x[i, 0] * x[i+1, 0] < 0:  # this changes the sign for the position to generate the data
        poincare_x.append(x[i+1, 0])
        poincare_v.append(x[i+1, 1])

# We now generate the graph to show the Poincar Section generated!
plt.figure(figsize=(8, 6))
plt.plot(poincare_x, poincare_v, 'bo', markersize=1)
plt.xlabel('Position ($x$)')
plt.ylabel('Velocity ($v$)')
plt.title('Poincar Section for the Duffing Oscillator')
plt.text(-0.004, 0.3, r'$\ddot{x} + \delta \dot{x} + \beta x + \alpha x^3 = \gamma \cos(\omega t)$', fontsize=12, ha='center')
plt.grid(True)
plt.show()
```

```python
# We will use now the Monte Carlo Integration Technique to find the volume of a hemisphere of radius 1

# We import our libraries
from random import random
from math import sqrt

# We set the number of random points in unit cube
N = 1000000
count = 0  # Counter of number of points within sphere

# We set the next to generate a random point in the unit cube
for j in range(N):
    point = (random(), random(), random())
    
    # This checks if the point is inside the sphere (radius = 1)
    if point[0]**2 + point[1]**2 + point[2]**2 < 1: # x**2+y**2+z**2<1
        count += 1

# This is how we calculate the volume estimate
Answer = float(count) / float(N)

# We multiply by 4 to get the final volume of the hemisphere
Answer *= 4

# Print the volume of the hemisphere along with the uncertainty
print(f'The volume of a hemisphere of radius 1 is: {Answer:.4f} \u00B1 {4 * sqrt(N) / float(N):.4f}')
```

```python
"""
We are going to write a code to simulate the behavior of 2000 individuals over 200 steps in a one-dimensional random walk. 
Each individual can take a step forward or backward at each step, and the code records the position 
of each individual at each step to calculate the average position and root mean square (RMS) position of the group.

"""
# We import our libraries
from random import choice  
import numpy as np  
import matplotlib.pyplot as plt  #
from scipy.optimize import curve_fit  # We import curve_fit function from scipy.optimize module for curve fitting

# We define the function for curve fitting
def power(x, a, b):
    return a * x**b

# We set the number of steps and people
steps = 200  # Number of steps
boys = 2000  # Number of people

# We create arrays to store sums
x_sum = np.zeros(steps)  # Array to store sum of positions
x2_sum = np.zeros(steps)  # Array to store sum of squared positions

# We simulate random walks
for _ in range(boys):
    x = np.zeros(steps)  # Initialize positions array for each person
    for i in range(1, steps):
        step = choice([-1, 1])  # Choose a step forward or backward
        x[i] = x[i-1] + step  # Update the position
    x_sum += x  # Sum up the positions
    x2_sum += x**2  # Sum up the squared positions

# We calculate averages
x_avg = x_sum / boys  # Average positions
RMS = np.sqrt(x2_sum / boys)  # RMS positions

# We define the time variable
t = np.arange(steps)  # Time steps

# We plot the data
plt.plot(t, x_avg, 'b-', label='Average Position')  # Plot average positions
plt.plot(t, RMS, 'g-', label='RMS Position')  # Plot RMS positions

# This is how we do the curve fitting
popt, pcov = curve_fit(power, t, RMS) 
plt.plot(t, power(t, *popt), label=f'Curve Fit: $y(t) = {popt[0]:.3f} \\times t^{{{popt[1]:.3f}}}$') # We plot it
 
# We add Labels and legend to the graph
plt.xlabel('Time (Step number)') 
plt.ylabel('Average and RMS position (Steps)')  
plt.legend()  

# we Show the plot
plt.show()  
```

```python
# We will solve the Schrodinger Equation numerically for a particle in a one-dimensional box

# We import our libraries
import numpy as np
import matplotlib.pyplot as plt

# We set the problem parameters
L = 1.0  # Length of the box
N = 1000  # Number of discretization points
hbar = 1.0  # Reduced Planck constant to 1
m = 1.0  # Particle mass

# We set the space discretization
x = np.linspace(0, L, N)
dx = x[1] - x[0]

# This define the potential function (particle in a box)
def V(x):
    return 0 if 0 < x < L else np.inf

# This is the function to solve the Schrdinger equation
def solve_schrodinger(E):
    psi = np.zeros_like(x)
    psi[0] = 0
    psi[1] = dx
    for i in range(1, N - 1):
        psi[i + 1] = 2 * psi[i] - psi[i - 1] - 2 * dx**2 * (E - V(x[i])) * psi[i]
    return psi

# We solve the Schrdinger equation by
E = float(input('Enter the energy: ')) # Test energy
psi_solution = solve_schrodinger(E)

# Plotting the wavefunction
plt.plot(x, psi_solution, label="Wavefunction")
plt.xlabel(r"$x$")
plt.ylabel(r"$\Psi(x)$")
plt.title("Particle in a one-dimensional box")
plt.legend()
plt.grid(True)
plt.show()
```
