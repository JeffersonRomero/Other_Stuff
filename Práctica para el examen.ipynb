Ejercicio 2: Crea una función llamada factorial que calcule el factorial de un número entero no negativo. El factorial de un número n (denotado como n!) es el producto de todos los enteros positivos menores o iguales n. Por ejemplo, 5!=5×4×3×2×1=120

Instrucciones: (1) La función debe recibir un número entero como parámetro. (2) Devuelve el factorial de ese número. (3) Si el número es negativo, la función debería devolver un mensaje de error o simplemente None.

```python
# Mi código
def factorial(xx):
    try:
        if xx>=1:
            ii=1
            factor=1
            while xx>0:
                xx=int(xx)
                factor=factor*xx
                xx=xx-ii
            return factor
        elif xx==0:
            return 1
        else:
            print('It has to be a positive integer')
    except ValueError:
        print('Wrong input!')
        
factorial(5)
```

```python
# Preferencia de la máquina
def factorial(xx):
    try:
        xx = int(xx)
        if xx < 0:
            print("It has to be a positive integer")
            return None
        elif xx == 0:
            return 1
        else:
            factor = 1
            for i in range(1, xx + 1):
                factor *= i
            return factor
    except ValueError:
        print("Wrong input!")
        
print(factorial(5))
```

Ejercicio 3: Crea una función llamada contar_vocales que tome una cadena de texto como parámetro y devuelva la cantidad de vocales que contiene (tanto minúsculas como mayúsculas).

Instrucciones: (1) La función debe recibir una cadena de texto. (2) Cuenta las vocales (a, e, i, o, u), sin importar si están en mayúscula o minúscula. (3) Devuelve el total de vocales en la cadena.

```python
# Mi código
vocales=['A','Á','a','á', 'E','É','e','é','I','Í','i','í', 'O','Ó','o','ó', 'U','Ú','u','ú']
def contar_vocales(xx):
    ii=0
    jj=0
    for ii in range(len(xx)):
        if xx[ii] in vocales:
            jj=jj+1
            ii=ii+1
        else:
            ii=ii+1
    return jj

contar_vocales('Jefferson Alexander Romero Márquez')
```

```python
# Preferencia de la máquina
def contar_vocales(xx):
    # Convertimos la cadena a minúsculas
    xx = xx.lower()
    # Definimos las vocales, incluyendo acentuadas
    vocales = "aeiouáéíóú"
    # Contador de vocales
    contador = 0
    # Recorremos cada carácter de la cadena
    for caracter in xx:
        if caracter in vocales:
            contador += 1
    return contador

# Prueba de la función
print(contar_vocales("Jefferson Alexander Romero Márquez"))  # Debería mostrar 13
```

Ejercicio 4: Crea una función llamada palindromo que determine si una cadena de texto es un palíndromo. Un palíndromo es una palabra o frase que se lee igual de izquierda a derecha que de derecha a izquierda, ignorando espacios, signos de puntuación y diferencias entre mayúsculas y minúsculas.

Instrucciones: (1) La función debe recibir una cadena de texto. (2) Ignora los espacios, los signos de puntuación y las mayúsculas. (3) Devuelve True si la cadena es un palíndromo o False en caso contrario.

```python
# Mi código
def palindromo(xx):
    xx=xx.replace(' ','')
    xx=xx.replace(".","")
    xx=xx.replace(',','')
    xx=xx.replace('!','')
    xx=xx.replace('¡','')
    xx=xx.replace(':','')
    xx=xx.replace(';','')
    xx=xx.replace("''",'')
    xx=xx.replace('?','')
    xx=xx.replace('¿','')
    xx=xx.replace('""','')
    if xx.lower()==xx[::-1].lower():
        return True
    else:
        return False
    
palindromo('.,Anita lava la tina!')
```

```python
# Preferencia de la máquina
import re

def palindromo(xx):
    # Convertimos la cadena a minúsculas y eliminamos todos los caracteres no alfabéticos
    xx = re.sub(r'[^a-záéíóúüñ]', '', xx.lower())
    # Comparamos la cadena con su reversa
    return xx == xx[::-1]

# Prueba de la función
print(palindromo("Anita lava la tina!"))  # Debería mostrar True
print(palindromo("Hola Mundo"))           # Debería mostrar False
```

Ejercicio 5: Crea una función llamada numeros_unicos que reciba una lista de números y devuelva otra lista con los números únicos en el mismo orden en que aparecen en la lista original. La lista resultante no debe contener números repetidos.

Instrucciones: (1) La función debe recibir una lista de números. (2) Debe devolver una lista nueva que contenga solo los números únicos, manteniendo su orden original. (3) No uses funciones de bibliotecas avanzadas (como set).

```python
# Mi código
def numeros_unicos(xx):
    ii=0
    lista=[]
    for ii in range(len(xx)):
        if xx[ii] in lista:
            ii=ii+1
        else:
            lista.append(xx[ii])
            ii=ii+1
    return lista

numeros_unicos([1,2,2,3,4,4,5])
```

```python
# Preferencia de la máquina
def numeros_unicos(xx):
    lista = []
    for numero in xx:
        if numero not in lista:
            lista.append(numero)
    return lista

# Prueba de la función
print(numeros_unicos([1, 2, 2, 3, 4, 4, 5]))  # Debería mostrar [1, 2, 3, 4, 5]
print(numeros_unicos([10, 20, 10, 30, 20]))   # Debería mostrar [10, 20, 30]
```

Ejercicio 4: Análisis de datos y filtro condicional.

Instrucciones: (1) Genera un DataFrame con dos columnas: (a) 'Estudiante': 50 nombres generados aleatoriamente a partir de una lista de nombres ('Ana', 'Luis', 'María', 'Juan', 'Carmen', etc.). (b) 'Nota': 50 calificaciones aleatorias entre 0 y 100. (2) Filtra el DataFrame para mostrar solo los estudiantes con una calificación superior a 70 (nota aprobatoria). (3) Calcula la calificación promedio de los estudiantes aprobados. (4) Visualiza la cantidad de estudiantes aprobados y reprobados en un gráfico de barras.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
estudiantes=random.choices(['Anne','Lois','Mary','Jack','Carmen'],k=50)
notas=np.random.randint(0,100,size=50)
df=pd.DataFrame({'Estudiante':estudiantes,'Nota':notas})
xx=df.where(df['Nota']>=70)
xx=xx.dropna()
promedio=xx['Nota'].mean()
print(f'El promedio de los estudiantes aprobados es: {promedio}')
yy=df.where(df['Nota']<70)
yy=yy.dropna()
aprobados=[]
aprobados.append(len(xx))
reprobados=[]
reprobados.append(len(yy))
df2=pd.DataFrame({'Aprobados':aprobados,'Reprobados':reprobados})
df2.plot(kind='bar')
plt.title('Estudiantes Aprobados')
plt.xlabel('')
plt.ylabel('Número')
plt.show()
```

```python
# Preferencia de la máquina
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

# Generar datos de ejemplo
estudiantes = random.choices(['Ana', 'Luis', 'María', 'Juan', 'Carmen'], k=50)
notas = np.random.randint(0, 100, size=50)

# Crear el DataFrame
df = pd.DataFrame({'Estudiante': estudiantes, 'Nota': notas})

# Filtrar estudiantes aprobados
aprobados = df[df['Nota'] > 70]
reprobados = df[df['Nota'] <= 70]

# Calcular el promedio de notas de estudiantes aprobados
promedio_aprobados = aprobados['Nota'].mean()
print(f"El promedio de los estudiantes aprobados es: {promedio_aprobados:.2f}")

# Contar el número de aprobados y reprobados
conteo = {'Aprobados': len(aprobados), 'Reprobados': len(reprobados)}

# Crear el gráfico de barras
plt.bar(conteo.keys(), conteo.values(), color=['#4C72B0', '#C44E52'])
plt.title('Estudiantes Aprobados y Reprobados')
plt.ylabel('Número')
plt.show()
```

Instrucciones: (1) Genera un conjunto de datos de temperaturas diarias: (a) Crea una lista de 30 fechas consecutivas a partir de hoy utilizando pandas.date_range. (b) Genera dos series de temperaturas aleatorias entre 15°C y 35°C, una para la ciudad de Madrid y otra para la ciudad de Barcelona. (2) Crea un DataFrame que contenga tres columnas: 'Fecha', 'Madrid' y 'Barcelona'. (3) Calcula el promedio de temperatura para cada ciudad. (4) Visualiza las temperaturas diarias de ambas ciudades en el mismo gráfico de línea con matplotlib. (5) Asegúrate de incluir etiquetas para cada ciudad, una leyenda y el título “Temperaturas Diarias de Madrid y Barcelona”.

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fechas=pd.date_range(start='4/11/2024',periods=30)
madrid=np.random.randint(15,35,size=30)
barcelona=np.random.randint(15,35,size=30)
df=pd.DataFrame({'Fecha':fechas,'Madrid':madrid,'Barcelona':barcelona})
promedio_madrid=round(df['Madrid'].mean(),2)
promedio_barcelona=round(df['Barcelona'].mean(),2)
print(f'La temperatura promedio para madrid es {promedio_madrid} y para Barcelona es {promedio_barcelona}')
plt.plot(df['Fecha'],df['Madrid'])
plt.plot(df['Fecha'],df['Barcelona'])
plt.legend(['Madrid','Barcelona'])
plt.title('Temperaturas Diarias de Madrid y Barcelona')
plt.show()
```

```python
#Preferencia de la máquina
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random

# Generar datos de ejemplo
fechas = pd.date_range(start='2024-11-04', periods=30)
madrid = np.random.randint(15, 35, size=30)
barcelona = np.random.randint(15, 35, size=30)

# Crear el DataFrame
df = pd.DataFrame({'Fecha': fechas, 'Madrid': madrid, 'Barcelona': barcelona})

# Calcular promedios
promedio_madrid = df['Madrid'].mean().round(2)
promedio_barcelona = df['Barcelona'].mean().round(2)

print(f"La temperatura promedio para Madrid es {promedio_madrid}°C y para Barcelona es {promedio_barcelona}°C")

# Graficar las series temporales
plt.plot(df['Fecha'], df['Madrid'], label='Madrid', color='blue')
plt.plot(df['Fecha'], df['Barcelona'], label='Barcelona', color='orange')
plt.title('Temperaturas Diarias de Madrid y Barcelona')
plt.xlabel('Fecha')
plt.ylabel('Temperatura (°C)')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()  # Ajusta el diseño para que las etiquetas no se solapen
plt.show()
```
