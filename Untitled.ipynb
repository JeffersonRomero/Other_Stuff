```python
import matplotlib.pyplot as plt

# Configuración del gráfico
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_xlim(-10, 120)
ax.set_ylim(-40, 40)
ax.set_aspect('equal')
ax.axhline(0, color='black', linewidth=0.8, linestyle='--')

# Representación de las lentes
lente_objetivo = 0
lente_intermedia = 50
ocular = 100

# Dibujar las lentes
ax.plot([lente_objetivo, lente_objetivo], [-20, 20], color='blue', linewidth=2, label="Objetivo")
ax.plot([lente_intermedia, lente_intermedia], [-15, 15], color='green', linewidth=2, label="Lente Intermedia")
ax.plot([ocular, ocular], [-10, 10], color='red', linewidth=2, label="Ocular")

# Dibujar los rayos desde el objetivo
rayos = [
    {'x': [lente_objetivo, lente_intermedia], 'y': [0, 10], 'color': 'orange'},  # Rayo central
    {'x': [lente_objetivo, lente_intermedia], 'y': [20, 15], 'color': 'purple'},  # Rayo superior
    {'x': [lente_objetivo, lente_intermedia], 'y': [-20, -15], 'color': 'purple'},  # Rayo inferior
]

# Rayos desde la lente intermedia al ocular
rayos_intermedia = [
    {'x': [lente_intermedia, ocular], 'y': [10, 5], 'color': 'orange'},  # Rayo central
    {'x': [lente_intermedia, ocular], 'y': [15, 10], 'color': 'purple'},  # Rayo superior
    {'x': [lente_intermedia, ocular], 'y': [-15, -10], 'color': 'purple'},  # Rayo inferior
]

# Dibujar rayos iniciales
for rayo in rayos:
    ax.plot(rayo['x'], rayo['y'], color=rayo['color'], linestyle='--')

# Dibujar rayos después de la lente intermedia
for rayo in rayos_intermedia:
    ax.plot(rayo['x'], rayo['y'], color=rayo['color'], linestyle='--')

# Etiquetas y leyenda
ax.text(lente_objetivo, 25, "Objetivo", color='blue', fontsize=10, ha='center')
ax.text(lente_intermedia, 20, "Lente Intermedia", color='green', fontsize=10, ha='center')
ax.text(ocular, 15, "Ocular", color='red', fontsize=10, ha='center')
ax.legend(loc='upper right')
ax.set_title("Esquema de los rayos en el sistema óptico")
ax.set_xlabel("Distancia (mm)")
ax.set_ylabel("Altura (mm)")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Distancias y alturas
dist_objetivo_lente_intermedia = 50  # mm
dist_lente_intermedia_ocular = 50  # mm
altura_maxima_objetivo = 30  # mm
altura_minima_objetivo = -30  # mm

# Coordenadas para los elementos
objetivo_x = [0, 0]
objetivo_y = [altura_minima_objetivo, altura_maxima_objetivo]

lente_intermedia_x = [dist_objetivo_lente_intermedia] * 2
lente_intermedia_y = [altura_minima_objetivo / 2, altura_maxima_objetivo / 2]

ocular_x = [dist_objetivo_lente_intermedia + dist_lente_intermedia_ocular] * 2
ocular_y = [altura_minima_objetivo / 4, altura_maxima_objetivo / 4]

# Rayos
rayos_x = [0, dist_objetivo_lente_intermedia, dist_objetivo_lente_intermedia + dist_lente_intermedia_ocular]
rayos_y_superior = [altura_maxima_objetivo, altura_maxima_objetivo / 2, altura_maxima_objetivo / 4]
rayos_y_inferior = [altura_minima_objetivo, altura_minima_objetivo / 2, altura_minima_objetivo / 4]

# Cálculo de los ángulos
angulo_wL = np.arctan((altura_maxima_objetivo - altura_maxima_objetivo / 2) / dist_objetivo_lente_intermedia) * 180 / np.pi
angulo_woc = np.arctan((altura_maxima_objetivo / 2 - altura_maxima_objetivo / 4) / dist_lente_intermedia_ocular) * 180 / np.pi

# Gráfico
plt.figure(figsize=(10, 6))

# Dibujar el objetivo
plt.plot(objetivo_x, objetivo_y, color="blue", label="Objetivo")

# Dibujar la lente intermedia
plt.plot(lente_intermedia_x, lente_intermedia_y, color="green", label="Lente Intermedia")

# Dibujar el ocular
plt.plot(ocular_x, ocular_y, color="red", label="Ocular")

# Dibujar los rayos
plt.plot(rayos_x, rayos_y_superior, color="orange", linestyle="--", label="Rayo Superior")
plt.plot(rayos_x, rayos_y_inferior, color="purple", linestyle="--", label="Rayo Inferior")

# Etiquetas de ángulos
plt.text(dist_objetivo_lente_intermedia / 2, altura_maxima_objetivo / 1.5,
         f"$w_L$ = {angulo_wL:.2f}°", color="orange", fontsize=10)
plt.text(dist_objetivo_lente_intermedia + dist_lente_intermedia_ocular / 2, altura_maxima_objetivo / 3,
         f"$w_{{oc}}$ = {angulo_woc:.2f}°", color="purple", fontsize=10)

# Configuración del gráfico
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Esquema de los rayos en el sistema óptico con ángulos")
plt.xlabel("Distancia (mm)")
plt.ylabel("Altura (mm)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Retry generating the plot with the same logic and details

import matplotlib.pyplot as plt
import numpy as np

# Distances and heights
dist_obj_lens = 50  # mm
dist_lens_ocular = 50  # mm
height_max_obj = 30  # mm
height_min_obj = -30  # mm

# Coordinates for elements
obj_x = [0, 0]
obj_y = [height_min_obj, height_max_obj]

lens_x = [dist_obj_lens] * 2
lens_y = [height_min_obj / 2, height_max_obj / 2]

ocular_x = [dist_obj_lens + dist_lens_ocular] * 2
ocular_y = [height_min_obj / 4, height_max_obj / 4]

# Rays
rays_x = [0, dist_obj_lens, dist_obj_lens + dist_lens_ocular]
rays_y_top = [height_max_obj, height_max_obj / 2, height_max_obj / 4]
rays_y_bottom = [height_min_obj, height_min_obj / 2, height_min_obj / 4]

# Calculate angles
angle_wL = np.arctan((height_max_obj - height_max_obj / 2) / dist_obj_lens) * 180 / np.pi
angle_woc = np.arctan((height_max_obj / 2 - height_max_obj / 4) / dist_lens_ocular) * 180 / np.pi

# Plot
plt.figure(figsize=(10, 6))

# Draw elements
plt.plot(obj_x, obj_y, color="blue", label="Objective")
plt.plot(lens_x, lens_y, color="green", label="Intermediate Lens")
plt.plot(ocular_x, ocular_y, color="red", label="Ocular")

# Draw rays
plt.plot(rays_x, rays_y_top, color="orange", linestyle="--", label="Top Ray")
plt.plot(rays_x, rays_y_bottom, color="purple", linestyle="--", label="Bottom Ray")

# Label angles
plt.text(dist_obj_lens / 2, height_max_obj / 1.5,
         f"$w_L$ = {angle_wL:.2f}°", color="orange", fontsize=10)
plt.text(dist_obj_lens + dist_lens_ocular / 2, height_max_obj / 3,
         f"$w_{{oc}}$ = {angle_woc:.2f}°", color="purple", fontsize=10)

# Configure plot
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.title("Optical System Ray Diagram with Angles")
plt.xlabel("Distance (mm)")
plt.ylabel("Height (mm)")
plt.legend()
plt.grid(True)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Posiciones de las lentes y elementos
pos_LE = 100  # Posición de la lucarna de entrada
pos_objetivo = 0  # Posición del objetivo
pos_intermedia = 100  # Posición de la lente intermedia
pos_ocular = 120  # Posición del ocular

# Dimensiones de los elementos
diam_LE = 60  # Diámetro de la LE (30 mm arriba y abajo)
diam_objetivo = 40  # Diámetro del objetivo (20 mm arriba y abajo)
diam_intermedia = 20  # Diámetro de la lente intermedia (10 mm arriba y abajo)
diam_ocular = 10  # Diámetro del ocular (5 mm arriba y abajo)

# Coordenadas de los elementos
LE = [(pos_LE, diam_LE / 2), (pos_LE, -diam_LE / 2)]
objetivo = [(pos_objetivo, diam_objetivo / 2), (pos_objetivo, -diam_objetivo / 2)]
intermedia = [(pos_intermedia, diam_intermedia / 2), (pos_intermedia, -diam_intermedia / 2)]
ocular = [(pos_ocular, diam_ocular / 2), (pos_ocular, -diam_ocular / 2)]

# Trayectorias de los rayos
rayo_pleno_superior = [
    (pos_LE, diam_LE / 2),  # Desde el borde superior de la LE
    (pos_objetivo, diam_objetivo / 2),  # Hasta el borde superior del objetivo
    (pos_intermedia, diam_intermedia / 2),  # Hasta el borde superior de la intermedia
    (pos_ocular, diam_ocular / 2)  # Hasta el borde superior del ocular
]

rayo_limite_superior = [
    (pos_LE, diam_LE / 2),  # Desde el borde superior de la LE
    (pos_intermedia, -diam_intermedia / 2),  # Hasta el borde inferior de la intermedia
    (pos_ocular, -diam_ocular / 2)  # Hasta el borde inferior del ocular
]

# Gráfico
plt.figure(figsize=(10, 6))

# Dibujo de los elementos
plt.plot([p[0] for p in LE], [p[1] for p in LE], label="Lucarna de Entrada (LE)", color="blue")
plt.plot([p[0] for p in objetivo], [p[1] for p in objetivo], label="Objetivo (Diafragma de Apertura)", color="black")
plt.plot([p[0] for p in intermedia], [p[1] for p in intermedia], label="Lente Intermedia", color="green")
plt.plot([p[0] for p in ocular], [p[1] for p in ocular], label="Ocular", color="red")

# Dibujo de los rayos
plt.plot([p[0] for p in rayo_pleno_superior], [p[1] for p in rayo_pleno_superior], '--', label="Rayo Pleno - Superior", color="orange")
plt.plot([p[0] for p in rayo_limite_superior], [p[1] for p in rayo_limite_superior], ':', label="Rayo Límite - Superior", color="purple")

# Configuración del gráfico
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.title("Esquema de los rayos desde el borde superior de LE con objetivo")
plt.xlabel("Distancia (mm)")
plt.ylabel("Altura (mm)")
plt.legend()
plt.grid()
plt.show()
```
